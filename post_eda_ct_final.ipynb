```python
def spl(oper, ind=0):
    try:
        s = oper.split('_')
        if ind == 0: return np.int16(s[0])
        if ind == 1: return np.int16(s[1])
    except:
        print("Некорректные данные")
        return None                         
```

Напишем функцию, которая булет проверять наличие всех значений конкретного признака тренировочной выборки в тестовой

```python
def check(col, X_full=X_full, X_test=X_test):
    a = set(X_test[col].unique()) - set(X_full[col].unique())
    b = set(X_full[col].unique()) - set(X_test[col].unique())
    if len(a) == 0: print(f'В тестовых категориях признака {col} есть все категории обучающей базы')
    else: print(f'В тестовой выборке есть занчение не встречающиеся в тренировочной {a}')
    return a
```

Почистим маленько текст в значениях этого столбца

```python
def clear_text(text):
    if text == '0': return 'missing'
    text = text.strip()
    text = text.partition(',')[0]
    text = text.partition('(')[0]
    text = text.partition('.')[0]
    text = text.partition('_')[0]
    text = text.partition('-')[0]
    text = re.sub(r'[^a-zA-Z ]', ' ', text).lower()
    #text = text.rstrip('s')
    if text == '' or text == ' ': text = 'missing'
    return " ".join(text.split())
```
